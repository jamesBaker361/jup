<u>Jarque-Bera Test for Normality</u>

The Jarque-Bera test statistic tests the null that the data is normally distributed against an alternative that the data follow some other distribution. The test statistic is based on two moments of the data, the skewness, and the kurtosis, and has an asymptotic $\chi^2_2$ distribution.

The test statistic is defined as

$JB = n \times \left( \dfrac{S^2}{6} + \dfrac{[K âˆ’ 3]^2}{24}\right)$

where $n$ is the number of data points, $S$ is the sample skewness, and $K$ is the sample kurtosis of the data.  The function returns:

1. JB value: The Jarque-Bera test statistic
2. JBpv p-value: The pvalue of the test statistic
3. Skew: Estimated skewness of the data
4. Kurtosis: Estimated kurtosis of the data

The skewness for a normal distribution is zero, and any symmetric data should have a skewness near zero. Negative values for the skewness indicate data that are skewed left and positive values for the skewness indicate data that are skewed right.

Kurtosis is a measure of whether the data are heavy-tailed or light-tailed relative to a normal distribution. That is, data sets with high kurtosis tend to have heavy tails, or outliers. Data sets with low kurtosis tend to have light tails, or lack of outliers. A uniform distribution would be the extreme case. The kurtosis for a standard normal distribution is three.


```python
##
## do JB test
##
import statsmodels.stats.stattools as smt
jb_results = smt.jarque_bera(reg01.resid)
print( 'JB Value: ', str( round( jb_results[ 0 ], 4 ) ) )
print( 'JB p-value: ', str( round( jb_results[ 1 ], 3 ) ) )
print( 'JB Skewness: ', str( round( jb_results[ 2 ], 3 ) ) )
print( 'JB Kurtosis: ', str( round( jb_results[ 3 ], 3 ) ) )
```

## <font color = blue>Partial Regression Plots</font>

Since we are doing multivariate regressions, we cannot just look at individual bivariate plots to discern relationships.
Instead, we want to look at the relationship of the dependent variable and independent variables conditional on the other
independent variables. We can do this through using partial regression plots, otherwise known as added variable plots.

In a partial regression plot, to discern the relationship between the response variable and the $k^{th}$ variabe, we compute
the residuals by regressing the response variable versus the independent variables excluding $X_k$. We can denote this by
$X_{âˆ¼k}$. We then compute the residuals by regressing $X_k$ on $X_{âˆ¼k}$. The partial regression plot is the plot
of the former versus the latter residuals.

The notable points of this plot are that the fitted line has slope $\beta_k$ and intercept zero. The residuals of this plot
are the same as those of the least squares fit of the original model with full $X$. You can discern the effects of the
individual data values on the estimation of a coefficient easily. If obs_labels is True, then these points are annotated
with their observation label. You can also see the violation of underlying assumptions such as homooskedasticity and
linearity.

```python
fig, ax = plt.subplots( figsize = ( 12, 8 ) )
fig = sm.graphics.plot_partregress( "prestige", "income", [ "income", "education" ], data = df, ax = ax )
```

```python
fix, ax = plt.subplots( figsize = ( 12, 14 ) )
fig = sm.graphics.plot_partregress( "prestige", "income", [ "education" ], data = df, ax = ax )
```

As you can see the partial regression plot confirms the influence of conductor, minister, and RR.engineer on the partial relationship between income and prestige. The cases greatly decrease the effect of income on prestige. Dropping these cases confirms this.
