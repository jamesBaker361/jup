## <font color = blue>Example 2: Patents</font>

### Data Dictionary

| Variable | Values   | Source | Mnemonic |
|----------|----------|--------|---------|
| Annual U.S. Patents | Number of patent applications filed, in thousands (Range 84.5 - 189.4) | Statistical Abstract of the U.S., various years | patents |
| R&D Expenditures | billions of 1992 dollars obtained as the ratio of expenditure on current dollars divided by the GDP price deflator.  (Range 57.94 - 166.7) | 1995 Statistical Abstract of the U.S., Table 979, Page 611, and earlier issues | rnd |


### Testable Hypothesis and Statistical Hyptheses

A amjor determinant of the number of patents is the amount of R&D expenditures.  R&D lays the groundwork for patents by providing the conditions and framework for patents.  It is expected that a positive relationship exists between R&D expenditures and patents because of these conditions and framework established by the R&D.

The model is:

$patents = \beta_0 + \beta_1 R\&D + \epsilon$

The statistical hypotheses are:

$H_O: \beta_1 = 0$

$H_A: \beta_1 > 0$

### Basic Data Loading and Examination

```python
##
## import packages
##
import numpy as np
import pandas as pd
##
## for modeling
## notice the new import command for the formula API and the summary option
##
import statsmodels.api as sm
import statsmodels.formula.api as smf 
from statsmodels.iolib.summary2 import summary_col
##
## for plotting
##
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##
## import patent data from a csv file
## notice the definition of a path variable
##
path = r'C:/Users/Owner/Documents/Teaching/Rutgers/ECMT322/labs/Data/'
df = pd.read_csv( path + 'patents.csv', index_col = 'year' )
df = df.drop( columns = [ 'obs' ] )
df.tail()
```

```python
##
## get basic descriptive statistics
##
df.describe().T
```

### OLS

```python
## 
## ols
##
## there are four steps for estimatng a model:
##   1. define a formula (i.e., the specific model to estimate)
##   2. instantiate the model (i.e., specify it)
##   3. fit the model
##   4. summarize the fitted model
##
## Step 1: define a formula
##
formula = 'patents ~ rnd'
##
## Step 2: instantiate the ols model
##
mod = smf.ols( formula, data = df )
##
## Step 3: fit the instantiated model
##
reg01 = mod.fit()
##
## Step 4: summarize the fitted model
##
print( reg01.summary() )
```

### OLS

```python
## 
## ols
##
## there are four steps for estimatng a model:
##   1. define a formula (i.e., the specific model to estimate)
##   2. instantiate the model (i.e., specify it)
##   3. fit the model
##   4. summarize the fitted model
##
## Step 1: define a formula
##
formula = 'rendcyco ~ rendmark'
##
## Step 2: instantiate the ols model
##
mod = smf.ols( formula, data = df )
##
## Step 3: fit the instantiated model
##
reg01 = mod.fit()
##
## Step 4: summarize the fitted model
##
print( reg01.summary() )
```

### Test CAPM Beta = 1

```python
##
## do f-test of estimated coefficient for rendmark = 1
##
## specify hypothesis
##
hypothesis = '(rendmark = 0)'
##
## run and print test 
##
f_test = reg01.f_test(hypothesis)
print(f_test)
print( "t-stat = ", round( np.sqrt( 241.33634292 ), 3 ) )
```
